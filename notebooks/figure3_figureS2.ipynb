Evaluate the 10-day forecast skill for the ML-corrected ("prognostic") simulations. This notebook depends on diagnostics computed by the `prognostic-reports` workflow.

Figure 3 and Figure S2 are generated by this notebook.

```python
import xarray as xr
import fsspec
import os
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import backoff
matplotlib.rcParams['figure.dpi'] = 300
xr.set_options(keep_attrs=True)
```

```python
root_url = 'gs://vcm-ml-archive/prognostic_run_diags'
runs = ['baseline', 'rf-control', 'rf-dQ1-dQ2-only']
cache_keys = {
    'baseline': 'vcm-ml-experiments-2020-10-30-nudge-to-obs-GRL-paper-weather-forecasts',
    'rf-control': 'vcm-ml-experiments-2021-03-15-nudge-to-obs-GRL-paper-rev1-weather-forecasts',
    'rf-dQ1-dQ2-only': 'vcm-ml-experiments-2021-03-15-nudge-to-obs-GRL-paper-rev1-weather-forecasts',
}
run_labels = {
    'baseline': 'baseline',
    'rf-control': 'ML-corrected',
    'rf-dQ1-dQ2-only': 'ML-corrected\n($\Delta Q_1$, $\Delta Q_2$ only)'
}
start_times = [f'2016{m:02}01.000000' for m in range(1, 13)]
```

```python
# for some reason gcsfs is sometimes unpredictably raising NotImplementedError when
# doing md5 checksum during loading of dataset
@backoff.on_exception(backoff.expo, NotImplementedError)
def load_dataset(url):
    with fsspec.open(diag_url) as f:
        ds = xr.open_dataset(f).load()
    return ds


def plot_lines(mean, std, variables, ylabel=None, legend_fontsize=10):
    nrow = int(np.ceil(len(variables)/3))
    fig, axes = plt.subplots(nrow, 3, figsize=(10, 2.6 * nrow))
    axes = axes.flatten()
    colors = ['k', 'r', 'b', 'm']
    for j, name in enumerate(variables):
        ax = axes[j]
        da_mean = mean[name]
        da_std = std[name]
        x = da_mean.time
        for i, exp in enumerate(da_mean.experiment):
            exp = exp.values.item()
            y = da_mean.sel(experiment=exp)
            spread = da_std.sel(experiment=exp)
            ax.plot(x, y, color=colors[i], linewidth=0.8, label=exp)
            ax.fill_between(x, y - spread, y + spread, color=colors[i], alpha=0.1)
        ax.set_title(da_mean.attrs['long_name'])
        ax.set_xlim([0, 10])
        ax.set_xlabel('time elapsed [days]')
        ax.set_xticks(range(0, 11))
        if ylabel:
            ax.set_ylabel(f"{ylabel} [{da_mean.units}]")
        else:
            ax.set_ylabel(f"{da_mean.units}")
        ax.grid()
        if j == 0:
            ax.legend(fontsize=legend_fontsize)
    fig.tight_layout(pad=0.1)
```

### Load diagnostics

```python
diags = []
for run in runs:
    run_diags = []
    for time in start_times:
        diag_url = os.path.join(root_url, f'{cache_keys[run]}-{time}_{run}', 'diags.nc')
        print(f"Loading {diag_url}")
        ds = load_dataset(diag_url)
        dt = ds.time.values[1] - ds.time.values[0]
        time_values = [t.days * 86400 + t.seconds for t in ds.time.values - ds.time.values[0]]
        new_time_coord = xr.DataArray(time_values, name="time")
        new_time_coord.values[:] = new_time_coord.values[:] + dt.seconds
        new_time_coord = (new_time_coord / 3600.0 / 24).assign_attrs(units="days")
        ds = ds.assign_coords(time=new_time_coord.values)
        run_diags.append(ds.assign_coords(start_time=time))
    run_diags = xr.concat(run_diags, dim='start_time').assign_coords(experiment=run_labels[run])
    diags.append(run_diags)
diags = xr.concat(diags, dim='experiment')
```

```python
diags_mean = diags.mean('start_time')
diags_std = diags.std('start_time')
```

### Plot RMSE errors averaged over 12 start dates
Shading shows +/- one standard deviation

### Figure 3

```python
paper_vars = ['h500_rms_global', 'pressfc_rms_global', 'tmplowest_rms_global']
plot_runs = ['baseline', 'ML-corrected']
# just baseline and rf-control
plot_lines(diags_mean.sel(experiment=plot_runs), diags_std.sel(experiment=plot_runs), paper_vars, ylabel='RMSE')
```

### Figure S3

```python
# all three experiments
plot_lines(diags_mean, diags_std, paper_vars, ylabel='RMSE', legend_fontsize=9)
```

Check some other variables out of interest.

```python
other_vars = ['pwat_rms_global', 'tmp850_rms_global', 'tmp200_rms_global', 'tmp500_300_rms_global', 'vgrd200_rms_global', 'ugrd200_rms_global']
plot_lines(diags_mean, diags_std, other_vars, ylabel='RMSE', legend_fontsize=9)
```

Plot difference in RMSE for "ML corrected" and "baseline cases" for all 12 forecasts. Negative values mean the ML-corrected forecast has a lower RMSE than the baseline forecast.

```python
relative_diff = diags.sel(experiment='ML-corrected') - diags.sel(experiment='baseline')
fig, axes = plt.subplots(1, 3)
for i, var in enumerate(paper_vars):
    relative_diff[var].plot(ax=axes[i], hue='start_time', color='black', alpha=0.5, linewidth=1, add_legend=False)
    axes[i].hlines(0, 0, 10, linewidth=0.8)
fig.set_size_inches(9, 2.5)
fig.tight_layout()
```
